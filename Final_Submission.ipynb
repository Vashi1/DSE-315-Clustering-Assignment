```python
import pandas as pd
from sklearn.cluster import BisectingKMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv', header=None)
data.head()
```

```python
plt.scatter(data[0], data[1])
plt.title("Scatter Plot for the Original Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Using Bisecting KMeans Clustering with Init = k-means++** 

```python
bkmp = BisectingKMeans(n_clusters=3, init='k-means++', random_state=42)
bkmp.fit(data)
```

```python
labels = bkmp.labels_
```

```python
plt.scatter(data[0], data[1], c = labels)
plt.title("Scatter Plot with Labeled Clusters")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
print(len(labels))
df = pd.DataFrame(data = labels)
```

```python
df.to_csv(r'final_labels.txt', header=None, index=None, sep=' ')
```
